```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Docking results for each ligand
data = {
    'Ligand': ['azaperone', 'azaperone', 'azaperone',
               'droperidol', 'droperidol', 'droperidol',
               'raclopride', 'raclopride', 'raclopride',
               'risperidone', 'risperidone', 'risperidone',
               'tiapride', 'tiapride', 'tiapride'],
    'Mode': [1, 2, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3],
    'RMSD (lower bound)': [0, 5.124, 3.219, 
                           0, 2.001, 8.924, 
                           0, 4.000, 2.543, 
                           0, 3.764, 5.123, 
                           0, 2.341, 7.123],
    'RMSD (upper bound)': [0, 9.502, 3.424, 
                           0, 2.539, 10.11, 
                           0, 4.24, 2.754, 
                           0, 6.421, 8.245, 
                           0, 5.654, 9.142]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Scatter plot for RMSD lower and upper bounds
plt.scatter(df['Mode'], df['RMSD (lower bound)'], color='blue', label='RMSD (Lower Bound)', marker='o')
plt.scatter(df['Mode'], df['RMSD (upper bound)'], color='red', label='RMSD (Upper Bound)', marker='x')

# Adding annotations for the ligands
for i in range(len(df)):
    plt.text(df['Mode'][i] + 0.1, df['RMSD (upper bound)'][i], df['Ligand'][i], fontsize=9)

# Customize the plot
plt.title('RMSD Scatter Plot for Azaperone, Droperidol, Raclopride, Risperidone, and Tiapride')
plt.xlabel('Docking Mode')
plt.ylabel('RMSD')
plt.legend()

# Show the plot
plt.show()
```
